```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

xp = np.array([10, 20, 40, 50, 90, 30, 70])
yp = np.array([1, 32, 42, 55, 75, 95, 45])

font1 = {"family":"serif", "color":"k", "size":30}
font2 = {"family":"arial", "color":"k", "size":20}

fig = plt.figure(figsize=(10, 6), facecolor="g", edgecolor="darkred", linewidth=5)
xy = fig.add_axes([1,1,1,1])
xy.plot(xp, yp, "h-c", ms="15", mfc="darkred", mec="darkred")


plt.title("Sports Watch Data", loc = "left", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.xticks(xp, labels=["1", "2", "3", "4", "5", "6", "7"])
plt.legend(["Linear Graph"])
plt.show()
```
