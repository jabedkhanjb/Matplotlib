# Data Visualization Linear Graph

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# Data in array
xpoint = np.array([1, 2, 5, 6, 7, 9, 10])
ypoint = np.array([2, 7, 3, 5, 9, 15, 20])


outerdesign = plt.figure(figsize=(10, 5), facecolor = "g", edgecolor = "m", linewidth = 3)
plotting = outerdesign.add_axes([1,1,1,1])

plt.title("Data Visualization", fontsize = 20, color = "k")
plt.xlabel("X-Axis", fontsize = 15, color = "k")
plt.ylabel("Y-Axis", fontsize = 15, color = "k")
plt.xticks(xpoint, labels = ["One", "Two", "Three", "Four", "Five", "Six", "Seven"])

plotting.plot(xpoint, ypoint, "h-c", ms = 10, mfc="m", mec="k")
plt.legend(["Linear Graph"])
plt.show()
```
