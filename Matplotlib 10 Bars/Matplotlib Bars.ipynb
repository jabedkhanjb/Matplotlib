# Matplotlib Bars

### Creating Bars

With Pyplot, you can use the <b>bar()</b> function to draw bar graphs:

```python
import matplotlib.pyplot as plt
import numpy as np

xaxis = np.array([2, 4, 6, 10, 11, 15])
yaxis = np.array([1, 4, 5, 8, 14, 10])

plt.xticks(xaxis, labels = ["A","B","C","D","E","F"])
plt.bar(xaxis, yaxis, color="m")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["APPLES", "BANANAS"]
y = [400, 350]

plt.bar(x, y, color="c")
plt.show()
```

# Horizontal Bars

If you want the bars to be displayed horizontally instead of vertically, use the <b>barh()</b> function:

<b>Draw n number of horizontal bars:</b>

```python
import matplotlib.pyplot as plt
import numpy as np

xaxis = np.array([2, 4, 6, 10, 11, 15])
yaxis = np.array([1, 4, 5, 8, 14, 10])

plt.xticks(xaxis, labels = ["A","B","C","D","E","F"])
plt.barh(xaxis, yaxis, color="m")
plt.show()
```

# Bar Width

The <b>bar()</b> takes the keyword argument <b>width</b> to set the width of the bars:

```python
import matplotlib.pyplot as plt
import numpy as np

xaxis = np.array([2, 4, 6])
yaxis = np.array([8, 14, 10])

plt.yticks(yaxis, labels = ["A","B","C"])
plt.bar(xaxis, yaxis, color="m", width = 0.5)
plt.show()
```

# Bar Height

The <b>barh()</b> takes the keyword argument <b>height</b> to set the height of the bars:

<i><b>Draw 3 very thin bars:</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np

xaxis = np.array([2, 4, 6])
yaxis = np.array([8, 14, 10])

plt.yticks(yaxis, labels = ["A","B","C"])
plt.barh(xaxis, yaxis, color="m", height = 0.5)
plt.show()
```
