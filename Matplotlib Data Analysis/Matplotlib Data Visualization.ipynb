# Setting Limits and Tick labels

<b>xlim()</b> and <b>ylim()</b> functions are used to set the limits of the X-axis and Y-axis respectively. Similarly, <b>xticks()</b> and <b>yticks()</b> functions are used to set tick labels.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([10, 15, 20, 30, 35, 40])
ypoint = np.array([20, 25, 30, 50, 52, 35])

plt.plot(xpoint, ypoint, "h-k", ms="15", mfc="m", mec="c")

plt.title("Project Class Data Analysis", fontsize = "20", color = "#062a4c")
plt.xlabel("X-Axis Data", fontsize = 12, color="#859105")
plt.ylabel("Y-Axis Data", fontsize = 12, color="#859105")

# setting the limit of x,y axis
plt.ylim(15, 100)
plt.xlim(8, 42)

# setting the label of xaxis
plt.xticks(xpoint, labels=["one", "two", "three", "four", "five", "six"])
# plt.yticks(ypoint, labels=["one", "two", "three", "four", "five", "six"])


plt.show()
```

# Figure Class

Consider the figure class as the overall window or page on which everything is drawn. It is a top-level container that contains one or more axes. A figure can be created using the <b>figure()</b> method.

<i> <b>Syntax:</b>
    class matplotlib.figure.Figure(figsize=None, dpi=None, facecolor=None, edgecolor=None, linewidth=0.0, frameon=None, subplotpars=None, tight_layout=None, constrained_layout=None)
</i>

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
import numpy as np

xpoint = np.array([10, 15, 20, 30, 35, 40])
ypoint = np.array([20, 25, 30, 50, 52, 35])

# creating a new figure with width, height in inches with face color
# edge color and line width

fig = plt.figure(figsize=(7, 5), facecolor="g", edgecolor="c", linewidth="7")

# creating a new axes for the figure
ax = fig.add_axes([1,1,1,1])

# adding the data to be plotted


plt.plot(xpoint, ypoint, "h-k", ms="15", mfc="m", mec="c")

plt.title("Project Class Data Analysis", fontsize = "20", color = "#062a4c")
plt.xlabel("X-Axis Data", fontsize = 12, color="#859105")
plt.ylabel("Y-Axis Data", fontsize = 12, color="#859105")

# setting the limit of x,y axis
plt.ylim(15, 70)
plt.xlim(8, 42)

# setting the label of xaxis
plt.xticks(xpoint, labels=["one", "two", "three", "four", "five", "six"])
# plt.yticks(ypoint, labels=["one", "two", "three", "four", "five", "six"])

# adding legend
plt.legend(["@jabedkhanjb"])
plt.show()
```

# Axes Class

<i><b>Syntax:</b> axes([left, bottom, width, height])</i>

<b>Adding Title –</b> <i>ax.set_title()</i>

<b>Adding X Label and Y label –</b> <i>ax.set_xlabel(), ax.set_ylabel()</i>

<b>Setting Limits –</b> <i>ax.set_xlim(), ax.set_ylim()</i>

<b>Tick labels –</b> <i>ax.set_xticklabels(), ax.set_yticklabels()</i>

<b>Adding Legends –</b> <i>ax.legend()</i>

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
import numpy as np

p1 = np.array([10, 20, 30, 40])
p2 = np.array([20, 25, 35, 55])

jb = plt.figure(figsize = (6, 4), facecolor = "g")

# Adding the axes to the figure
fig = jb.add_axes([1,1,1,1])

# plotting 1st dataset to the figure
fig1 = fig.plot(p1, p2)

# plotting 2nd dataset to the figure
fig2 = fig.plot(p2, p1)


# setting title
fig.set_title("Linear Graph")

# setting label
fig.set_xlabel("X-Axis", fontsize = 15)
fig.set_ylabel("Y-Axis", fontsize = 15)


# plt.title("Linear Graph")
# plt.xlabel("X-Axis")
# plt.ylabel("Y-Axis")

# plt.legend(labels = ("Line 1", "Line 2"))
fig.legend(labels = ("Line 1", "Line 2"))
plt.show()
```

# Multiple Plots

<b>Syntax:</b> axes([left, bottom, width, height])

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
import numpy as np

p1 = np.array([10, 20, 30, 40])
p2 = np.array([20, 25, 35, 55])

jb = plt.figure(figsize = (5, 4), facecolor = "g", edgecolor = "darkred", linewidth = 2)
ax1 = jb.add_axes([0.1,0.1,0.8,0.8])
ax2 = jb.add_axes([1.1,0.1,0.8,0.8])

ax1.plot(p1, p2, "o--g", ms = 10)
ax1.plot(p2, p1, "o--m", ms = 10)
ax1.legend(labels=(["Green", "Megenda"]))

ax1.plot(p1, p2, "o--g", ms = 10)
ax1.plot(p2, p1, "o--m", ms = 10)
ax1.legend(labels=(["Green", "Megenda"]))

ax2.plot(p1, p2, "h-c", ms = 10)
ax2.plot(p2, p1, "h-y", ms = 10)
ax2.legend(labels=(["Cyan", "Yellow"]))

ax1.set_title("Linear Graph 1", fontsize = 20)
ax1.set_xlim(0, 60)
ax1.set_ylim(0, 70)
ax1.set_xlabel("X-Axis", fontsize = 15)
ax1.set_ylabel("Y-Axis", fontsize = 15)

ax2.set_title("Linear Graph 2", fontsize = 20)
ax2.set_xlim(0, 60)
ax2.set_ylim(0, 70)
ax2.set_xlabel("X-Axis", fontsize = 15)
ax2.set_ylabel("Y-Axis", fontsize = 15)


plt.show()
```

# Method 2: Using subplot() method.

This method adds another plot at the specified grid position in the current figure.

<b>Syntax:</b>

    subplot(nrows, ncols, index, **kwargs)

    subplot(pos, **kwargs)

    subplot(ax)

```python
import matplotlib.pyplot as plt
 
# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]
 
#  Creating figure object
plt.figure()
 
# adding first subplot
plt.subplot(121)
plt.plot(x, y)
 
# adding second subplot
plt.subplot(122)
plt.plot(y, x)
plt.show()
```

# Method 3: Using subplots() method

This function is used to create figures and multiple subplots at the same time.

<b>Syntax:</b>

    matplotlib.pyplot.subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)

```python
import matplotlib.pyplot as plt
import numpy as np
# initializing the data
x = np.array([10, 20, 30, 40])
y = np.array([20, 25, 35, 55])
 
#  Creating figure and subplots
fig, axes = plt.subplots(1, 2)

# plotting the data in the 1st subplot
axes[0].plot(x, y)
axes[0].plot(y, x)

 
# plotting the data in the 2nd subplot
axes[1].plot(x, y)
axes[1].plot(y, x)

plt.show()
```

# Method 4: Using subplot2grid() method

This function creates axes object at a specified location inside a grid and also helps in spanning the axes object across multiple rows or columns. In simpler words, this function is used to create multiple charts within the same figure.

<b>Syntax:</b>

    Plt.subplot2grid(shape, location, rowspan, colspan)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([10, 20, 30, 40])
y = np.array([20, 25, 35, 55])

# adding the subplots 
axes1 = plt.subplot2grid(
(7,1), (0,0), rowspan = 2, colspan = 1
)

axes2 = plt.subplot2grid(
(7,1), (2,0), rowspan = 2, colspan = 1
)

# plotting the data

axes1.plot(x, y)
axes2.plot(y, x)
plt.show()
```

# Different types of Matplotlib Plots

Matplotlib supports a variety of plots including line charts, bar charts, histograms, scatter plots, etc. We will discuss the most commonly used charts in this article with the help of some good examples and will also see how to customize each plot.  

Note: Some elements like axis, color are common to each plot whereas some elements are pot specific.
Line Chart

Line chart is one of the basic plots and can be created using the plot() function. It is used to represent a relationship between two data X and Y on a different axis.

<b>Syntax:</b>

    matplotlib.pyplot.plot(\*args, scalex=True, scaley=True, data=None, \*\*kwargs)

<b>Let’s see how to customize the above-created line chart. We will be using the following properties – </b>

    color: Changing the color of the line
    linewidth: Customizing the width of the line
    marker: For changing the style of actual plotted point
    markersize: For changing the size of the markers
    linestyle: For defining the style of the plotted line
