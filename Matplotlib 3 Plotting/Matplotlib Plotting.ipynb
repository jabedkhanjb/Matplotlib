# Plotting x and y points

The <b>plot()</b> function is used to draw points (markers) in a diagram.

By default, the <b>plot()</b> function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

<i><b>Draw a line in a diagram from position (1, 3) to position (8, 10):</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np

xaxis = np.array([1, 3])
yaxis = np.array([8, 10])

plt.plot(xaxis, yaxis)
plt.show()
```

# Plotting Without Line

To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

<i><b>Draw two points in the diagram, one at position (1, 3) and one in position (8, 10):</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np

xaxis = np.array([1, 3])
yaxis = np.array([8, 10])

plt.plot(xaxis, yaxis, "o") # simply use a "o" string to get dot
plt.show()
```

# Multiple Points

You can plot as many points as you like, just make sure you have the same number of points in both axis.

<i><b>Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):</b></i>

<i>Here position (1, 3) defines (x, y) and position (2, 8) is (x, y) and same for rest positions.</i>

```python
import matplotlib.pyplot as plt
import numpy as np

xaxis = np.array([1, 2, 6, 8])
yaxis = np.array([3, 8, 1, 10])

plt.plot(xaxis, yaxis)
plt.show()
```

# Default X-Points

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 (etc., depending on the length of the y-points.

So, if we take the same example as above, and leave out the x-points, the diagram will look like this:

<i><b>Plotting without x-points:</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array([5, 9, -2, 2, 10, 1, 7])
plt.plot(ypoint)
plt.show()
```
