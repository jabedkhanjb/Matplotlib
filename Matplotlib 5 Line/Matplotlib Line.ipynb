## Linestyle

You can use the keyword argument <b>linestyle</b>, or shorter <b>ls</b>, to change the style of the plotted line:

<i>Use a dotted line:</i>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array([2, 5, 1, 3, 9])

# dotted
plt.plot(ypoint, linestyle = "dotted")
plt.show()
```

<i>Use a dotted line:</i>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array([2, 5, 1, 3, 9])

# dotted
# dashed
plt.plot(ypoint, linestyle = "dashed")
plt.show()
```

# Shorter Syntax

The line style can be written in a shorter syntax:

<b>linestyle</b> can be written as <b>ls</b>.

<b>dotted</b> can be written as <b>:</b>.

<b>dashed</b> can be written as <b>--</b>.

<i>Shorter syntax:</i>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array([2, 5, 1, 3, 9])

# dotted
# dashed
plt.plot(ypoint, ls = ":") # linestyle as ls
plt.show()
```

# Line Styles

You can choose any of these styles:
Style 	Or

'solid' (default)       	<b>'-' </b>

'dotted' 	                <b>':' </b>

'dashed' 	                <b>'--' </b>

'dashdot' 	                <b>'-.' </b>

'None' 	                    <b>'' or ' '</b>

# Line Color

You can use the keyword argument <b>color</b> or the shorter <b>c</b> to set the color of the line:

<i>Set the line color to red:</i>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array([4, 2, 9, 12])

plt.plot(ypoint, color = "r")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array([4, 2, 1, 12])

plt.plot(ypoint, c = "hotpink")
plt.show()
```

# Line Width

You can use the keyword argument <b>linewidth</b> or the shorter <b>lw</b> to change the width of the line.

The value is a floating number, in points:

<i>Plot with a 20.5pt wide line:</i>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array([4, 1, 9, 7])

plt.plot(ypoint, "h--g", linewidth = "10", ms = 20) # can be used as int or str
plt.show()
```

# Multiple Lines

You can plot as many lines as you like by simply adding more <b>plt.plot()</b> functions:

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 9, 7, 2, 5])
y2 = np.array([5, 2, 1, 9, 3])
y3 = np.array([1, 3, 3, 5, 7])

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

plt.legend(["Linear Graph"])
plt.show()
```

We can also plot many lines by adding the points for the x- and y-axis for each line in the same <b>plt.plot()</b> function.

(In the examples above we only specified the points on the y-axis, meaning that the points on the x-axis got the the default values (0, 1, 2, 3).)

<i>Draw two lines by specifiyng the x- and y-point values for both lines:</i>

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)

plt.legend(labels = ("Blue", "Yellow"))
plt.show()
```
