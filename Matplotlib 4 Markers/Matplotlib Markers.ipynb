# Markers

You can use the keyword argument <i><b>marker</b></i> to emphasize each point with a specified marker:

<i><b>Mark each point with a circle:</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([0, 2, 4, 5, 7, 8])
ypoint = np.array([0, 45, 15, 150, 15, 15])

plt.plot(xpoint, ypoint, marker = "o")
plt.show()
```

<i><b>Mark each point with a star:</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([0, 2, 4, 5, 7, 8])
ypoint = np.array([0, 45, 15, 150, 15, 15])

plt.plot(xpoint, ypoint, marker = "*")
plt.show()
```

# Marker Reference

You can choose any of these markers:
Marker 	Description
'o' 	Circle 	
'*' 	Star 	
'.' 	Point 	
',' 	Pixel 	
'x' 	X 	
'X' 	X (filled) 	
'+' 	Plus 	
'P' 	Plus (filled) 	
's' 	Square 	
'D' 	Diamond 	
'd' 	Diamond (thin) 	
'p' 	Pentagon 	
'H' 	Hexagon 	
'h' 	Hexagon 	
'v' 	Triangle Down 	
'^' 	Triangle Up 	
'<' 	Triangle Left 	
'>' 	Triangle Right 	
'1' 	Tri Down 	
'2' 	Tri Up 	
'3' 	Tri Left 	
'4' 	Tri Right 	
'|' 	Vline 	
'_' 	Hline
