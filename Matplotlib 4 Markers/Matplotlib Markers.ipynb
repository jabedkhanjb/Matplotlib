# Markers

You can use the keyword argument <i><b>marker</b></i> to emphasize each point with a specified marker:

<i><b>Mark each point with a circle:</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([0, 2, 4, 5, 7, 8])
ypoint = np.array([0, 45, 15, 150, 15, 15])

plt.plot(xpoint, ypoint, marker = "o")
plt.show()
```

<i><b>Mark each point with a star:</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([0, 2, 4, 5, 7, 8])
ypoint = np.array([0, 45, 15, 150, 15, 15])

plt.plot(xpoint, ypoint, marker = "*")
plt.show()
```

# Format Strings <i>fmt</i>

You can use also use the shortcut string notation parameter to specify the marker.

This parameter is also called <b>fmt</b>, and is written with this syntax:

<i><b>marker | line | color</b></i>

<i><b>Mark each point with a circle:</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array([3, 8, 1, 10])
plt.plot(ypoint, "2-.k")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xpoint = np.array([0, 1.5, 2, 3, 4, 5, 6.5, 7.2, 8.5, 9.3, 10.5, 11.1, 12, 13, 14.7, 15.2])
ypoint = np.array([0, 15, 10, 15, 10, 25, 35, 20, 30, 40, 20, 10, 30, 35, 30, 10])

plt.plot(xpoint, ypoint, "p:g")
plt.show()
```

# Line Reference
Line Syntax 	Description

':' 	Dotted line 	
'--' 	Dashed line 	
'-.' 	Dashed/dotted line

<b>Note:</b> If you leave out the line value in the fmt parameter, no line will be plotted.

# Marker Size

You can use the keyword argument <i><b>markersize</b></i> or the shorter version, <i><b>ms</b></i> to set the size of the markers:

<i><b>Set the size of the markers to 20:</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, "D-g", ms = 10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

# Marker Color

You can use the keyword argument <b>markeredgecolor</b> or the shorter <b>mec</b> to set the color of the edge of the markers:

<i><b>Set the EDGE color to red:</b></i>

```python
import numpy as np
import matplotlib.pyplot as plt

xpoint = np.array([0, 1.5, 2, 3, 4, 5, 6.5, 7.2, 8.5, 9.3, 10.5, 11.1, 12, 13, 14.7, 15.2])
ypoint = np.array([0, 15, 10, 15, 10, 25, 35, 20, 30, 40, 20, 10, 30, 35, 30, 10])

plt.plot(xpoint, ypoint, "p-m", ms = "15", mec = "g")
plt.show()
```

We can use the keyword argument <i><b>markerfacecolor</b></i> or the shorter <i><b>mfc</b></i> to set the color inside the edge of the markers:

<i><b>Set the FACE color to red:</b></i>

```python
import numpy as np
import matplotlib.pyplot as plt

xpoint = np.array([0, 1.5, 2, 3, 4, 5, 6.5, 7.2, 8.5, 9.3, 10.5, 11.1, 12, 13, 14.7, 15.2])
ypoint = np.array([0, 15, 10, 15, 10, 25, 35, 20, 30, 40, 20, 10, 30, 35, 30, 10])

plt.plot(xpoint, ypoint, "D:c", ms = "10", mfc = "m")
plt.show()
```
