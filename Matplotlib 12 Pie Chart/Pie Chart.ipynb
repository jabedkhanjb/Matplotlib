# Matplotlib Pie Chart

### Creating Pie Charts

With Pyplot, we can use the <b>pie()</b> function to draw pie charts:

```python
import matplotlib.pyplot as plt
import numpy as np
import random

rany = random.sample(range(1, 100), 5)
y = np.array(rany)
mylabels = ["iPhone", "Android", "Windows", "Macbook", "Button"]
# pie() syntax
# plt.pie(y, labels = mylabels)

# colors = random.sample(range(1, 100), 5)
# print(colors)

plt.bar(mylabels, y, color = "m")
plt.show()

plt.pie(y, labels = mylabels)
plt.show()

```

# Start Angle

As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a <b>startangle</b> parameter.

The <b>startangle</b> parameter is defined with an angle in degrees, default angle is 0:

```python
import matplotlib.pyplot as plt
import numpy as np
import random

rany = random.sample(range(1, 100), 5)
y = np.array(rany)
mylabels = ["iPhone", "Android", "Windows", "Macbook", "Button"]

# startangle = "0, 90, 180, 270" can be anything
plt.pie(y, labels = mylabels, startangle = 150)
plt.show()
```

# Explode

Maybe we want one of the wedges to stand out? The <b>explode</b> parameter allows you to do that.

The <b>explode</b> parameter, if specified, and not <b>None</b>, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

<i><b>Pull any wedge 0.2 from the center of the pie:</b></i>

```python
import matplotlib.pyplot as plt
import numpy as np
import random

rany = random.sample(range(1, 100), 5)
y = np.array(rany)
mylabels = ["iPhone", "Android", "Windows", "Macbook", "Button"]

# explode
myexplode = [0.2, 0, 0, 0, 0]
plt.pie(y, labels = mylabels, startangle = 0, explode = myexplode)
plt.show()
```

# Shadow

Add a shadow to the pie chart by setting the <b>shadows</b> parameter to <b>True</b>:

```python
import matplotlib.pyplot as plt
import numpy as np
import random

rany = random.sample(range(1, 100), 5)
y = np.array(rany)
mylabels = ["iPhone", "Android", "Windows", "Macbook", "Button"]

# shadow = True
myexplode = [0.2, 0, 0, 0, 0]
plt.pie(y, labels = mylabels, startangle = 0, explode = myexplode, shadow = True)
plt.show()
```

# Colors

<i><b>syntax: colors</b></i>


```python
import matplotlib.pyplot as plt
import numpy as np
import random

rany = random.sample(range(1, 100), 5)
y = np.array(rany)
mylabels = ["iPhone", "Android", "Windows", "Macbook", "Button"]

# random color setting
ranc = random.sample(range(1, 100), 5)



mycolors = ["black", "hotpink", "b", "#4CAF50", "m", "c", "k", "g"]
ranc2 = random.sample(mycolors, 4)




# shadow = True
myexplode = [0.1, 0, 0, 0, 0]
plt.pie(y, colors = ranc2, labels = mylabels, startangle = 0, explode = myexplode, shadow = False)
plt.show()
```

# Legend

To add a list of explanation for each wedge, use the <b>legend()</b> function:

```python
import matplotlib.pyplot as plt
import numpy as np
import random

rany = random.sample(range(1, 100), 5)
y = np.array(rany)
mylabels = ["iPhone", "Android", "Windows", "Macbook", "Button"]

myexplode = [0.2, 0, 0, 0, 0]
plt.pie(y, labels = mylabels, shadow = False)

# legend
plt.legend()
plt.show()
```

# Legend With Header

To add a header to the legend, add the <b>title</b> parameter to the <b>legend</b> function.

```python
import matplotlib.pyplot as plt
import numpy as np
import random

rany = random.sample(range(1, 100), 5)
y = np.array(rany)
mylabels = ["iPhone", "Android", "Windows", "Macbook", "Button"]

myexplode = [0.2, 0, 0, 0, 0]
plt.pie(y, labels = mylabels, shadow = False)

# legend with title
plt.legend(title = "Customer Consumes")
plt.show()
```
