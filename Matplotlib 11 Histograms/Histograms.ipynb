# Create Histogram

In Matplotlib, <b>hist()</b> function is used to create histograms.

The <b>hist()</b> function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

```python
# normal data distribution

import numpy as np
import matplotlib.pyplot as plt


x = np.random.normal(10, 10, 50)
print(x)

plt.hist(x, color = "m")
plt.show()
```
