# Matplotlib Labels and Title

### Create Labels for a Plot

With Pyplot, you can use the <b>xlabel()</b> and <b>ylabel()</b> functions to set a label for the x- and y-axis.

<i>Add labels to the x- and y-axis:</i>

```python
import numpy as np
import matplotlib.pyplot as plt

xpoint = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
ypoint = np.array([40, 50, 160, 170, 280, 290, 300, 310, 320, 330])

plt.plot(xpoint, ypoint)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")


plt.show()
```

# Create a Title for a Plot

With Pyplot, you can use the <b>title()</b> function to set a title for the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

xpoint = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
ypoint = np.array([40, 50, 160, 170, 280, 290, 300, 310, 320, 330])

plt.plot(xpoint, ypoint, marker="p",c="m")

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.legend(["Heartrate"])

plt.show()
```

# Set Font Properties for Title and Labels

You can use the <b>fontdict</b> parameter in <b>xlabel()</b>, <b>ylabel()</b>, and <b>title()</b> to set font properties for the title and labels.

<i>Set font properties for the title and labels:</i>

```python
import numpy as np
import matplotlib.pyplot as plt

xpoint = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
ypoint = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# defining font style, color, size
font1 = {"family":"arial", "color":"green", "size":20}
font2 = {"family":"serif", "color":"darkred", "size":15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(xpoint, ypoint, marker = "p", mfc="m", mec="m")
plt.show()
```

# Position the Title

You can use the <b>loc</b> parameter in <b>title()</b> to position the title.

Legal values are: 'left', 'right', and 'center'. Default value is 'center'

```python
import numpy as np
import matplotlib.pyplot as plt

xpoint = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
ypoint = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# defining font style, color, size
font1 = {"family":"arial", "color":"green", "size":20}
font2 = {"family":"serif", "color":"darkred", "size":15}

plt.title("Sports Watch Data", loc = "left", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(xpoint, ypoint, marker = "p", mfc="m", mec="m")
plt.show()
```
